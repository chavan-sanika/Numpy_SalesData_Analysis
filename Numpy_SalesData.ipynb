# Sales Data Analysis

**1. Create Sales Data**

```python
sales = np.array([
 [200, 220, 250, 275, 300, 310], # Product A
 [150, 160, 170, 180, 190, 200], # Product B
 [100, 120, 130, 140, 150, 160], # Product C
 [90, 95, 100, 105, 110, 115] # Product D
])
products = ['Product A', 'Product B', 'Product C', 'Product D']
months = ['Jan', 'Feb', 'Mar', 'Apr', 'May', 'Jun']
```

```python
sales.shape
```

**2. Monthly & Product Analysis**

*Calculate total sales per product (row-wise sum)*

```python
total_sales = sales.sum(axis=1)
for i,j in zip(total_sales,products):
  print(f' total sales for {j} = {i} ')

total_sales
```

*Calculate total sales per month (column-wise sum).*

```python
total_sales_mon = sales.sum(axis=0)
for i,j in zip(total_sales_mon,months):
  print(f' total sales for {j} = {i}')
```

*Identify the best-selling product (based on total).*

```python
a =  total_sales.argmax()
print("Product with highest overall sales is:", products[a])
```

analysis - Product A had the highest total sales over six months, making it the best-selling product.

*Identify the month with the highest overall sales*

```python
b =  total_sales_mon.argmax()
print("Month with highest overall sales is:", months[b])
```

analysis - June recorded the highest overall sales across all products

**3.Percentage Growth**

*Calculate percentage growth from Jan to Jun for each product:*

*growth = ((sales[:, 5] - sales[:, 0]) / sales[:, 0]) * 100*

```python
growth = ((sales[:, 5] - sales[:, 0]) / sales[:, 0]) * 100
for i in range(len(products)):
    print(f"Percentage growth for {products[i]}: {growth[i]:.0f}%")
```

analysis -

 Product C showed the highest percentage growth from January to June i.e - 60%

 Product D had the lowest sales and growth


# Sales Visualization Dashboard


```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
plt.plot(months, sales[0], marker='o', color='orange', label='A')
plt.plot(months, sales[1], marker='s', color='green', label='B')
plt.plot(months, sales[2], marker='^', color='red', label='C')
plt.plot(months, sales[3], marker='d', color='black', label='D')

plt.title('Monthly Sales Graph')
plt.xlabel('Months')
plt.ylabel('Sales')

plt.legend()
plt.grid(True)

plt.show()
```

*Products sale in June month*

```python
june_sales = sales[:, 5]

plt.bar(products, june_sales, color=['orange', 'green', 'red', 'black'])

plt.title("Product-wise Sales in June")
plt.xlabel("Products")
plt.ylabel("Sales in June")
plt.grid(True)
plt.show()
```

```python
june_sales = pd.Series(sales[:, 5], index=products)
plt.figure(figsize=(5,4))
june_sales.plot.pie(autopct='%.0f%%', labels=june_sales.index)
```

analysis - we can see that Product A contributed the most to June sales
